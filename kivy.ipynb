Esse próximo seria para acrescentar vários widgets. Nesse caso, acrescentamos um botão e um label. Essa é uma das maneiras de se fazer isso. A outra maneira seria utilizando um arquivo .kv com a linguagem kivy junto com um arquivo python.

```python
from kivy.app import App
from kivy.uix.button import Button 
from kivy.uix.boxlayout import BoxLayout 
from kivy.uix.label import Label

class Test(App):
    def build(self):
        box = BoxLayout(orientation = 'vertical')
        button = Button(text='Botão 1')
        label = Label(text='Texto 1')
        box.add_widget(button)
        box.add_widget(label)
        
        box2 = BoxLayout()
        button2 = Button(text='Botão 2')
        label2 = Label(text='Texto 2')
        box2.add_widget(button2)
        box2.add_widget(label2)

        box.add_widget(box2)

        return box

Test().run()
```

No script abaixo estamos acrescentando o parâmetrop on_release no qual colocamos uma função.

```python
from kivy.app import App
from kivy.uix.button import Button 
from kivy.uix.boxlayout import BoxLayout 
#Existem vários layouts diferentes 
from kivy.uix.label import Label

class Test(App):
    def build(self):
        box = BoxLayout(orientation = 'vertical')
        button = Button(text='Botão 1',font_size = 30,on_release = self.incrementar) # Nesse on_release vamos colocar 
                                                                                     #uma função que será definida abaixo
        self.label = Label(text='1',font_size = 30)
        box.add_widget(button)
        box.add_widget(self.label)
   
        return box

    def incrementar(self,button):   #Toda vez que clicarem no botão , essa função será chamada
        #button.text = 'Soltei'   # Aqui a função muda o texto do botão assim que clicar e soltar
        self.label.text = str(int(self.label.text) + 1)



Test().run()
```

```python
from kivy.app import App
from kivy.uix.boxlayout import BoxLayout 

#Tiramos o ScrollView aqui porque ele  só deixa ter um filho. Aí daria problema na hora de acrescentar
#mais widgets dinamicamente. No arquivo .kv vai ter o Scroll view pra ver só as tarefas
class Tarefas(BoxLayout):
    def __init__(self,tarefas,**kwargs): #keywords arguments.Colocando o kwargs, nos permite 
                                         #acrescentar parâmetros para a função 

        super().__init__(**kwargs)   #Aqui estou garantindo que o método init do BoxLayout não seja sobreescrito
        for tarefa in tarefas:
            self.ids.box.add_widget(Tarefa(text = tarefa))
    def addWidget(self):
        #Aqui estou definindo que o texto vai ser o texto do text input, que definimos no arquivo
        #.kv como tendo id texto. E colocamos esse .text depois para capturarmos o texto escrito
        #e utiliz-alo na função
        texto = self.ids.texto.text 
        self.ids.box.add_widget(Tarefa(text = texto))
        #Aqui estou colocando para apagar o que a gente escreveu depois de clicar no botão
        self.ids.texto.text = ''

class Tarefa(BoxLayout):
    def __init__(self,text = '',**kwargs):
        super().__init__(**kwargs)
        self.ids.label.text = text        

class Test(App):
    def build(self):
        return Tarefas(['Fazer compras ', 'Buscar filho','Molhar a calçada','Dar comida pro cão','Beber agua'])
 
Test().run()
```
